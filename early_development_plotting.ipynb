```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
```

```python
acc_bottom=pd.read_csv("/Users/asmaimran/MBRU/AbdulRahman_study/early_development/enrichment_results/acc_bottom_25.csv")
acc_top=pd.read_csv("/Users/asmaimran/MBRU/AbdulRahman_study/early_development/enrichment_results/acc_top_75.csv")
```

```python
mtg_bottom=pd.read_csv("/Users/asmaimran/MBRU/AbdulRahman_study/early_development/enrichment_results/mtg_bottom_25.csv")
mtg_top=pd.read_csv("/Users/asmaimran/MBRU/AbdulRahman_study/early_development/enrichment_results/mtg_top_75.csv")
```

```python
visp_bottom=pd.read_csv("/Users/asmaimran/MBRU/AbdulRahman_study/early_development/enrichment_results/visp_bottom_25.csv")
visp_top=pd.read_csv("/Users/asmaimran/MBRU/AbdulRahman_study/early_development/enrichment_results/visp_top_75.csv")
```

```python
SUB = str.maketrans("0123456789", "")
plt.rc('xtick',labelsize=40)
plt.rc('ytick',labelsize=40)
fig, ax = plt.subplots(figsize=(15,15),dpi=600)
ax.xaxis.set_tick_params(length=10,pad=10)
ax.yaxis.set_tick_params(length=10,pad=10)
tick_spacing=1
plt.title("Replication timing - ACC", pad=30, fontsize=60, fontweight="bold")
ax.set_xlabel('-log10(P)'.translate(SUB), fontsize=50, labelpad=30, fontweight="bold")
ax.set_ylabel('Clusters', fontsize=50, labelpad=30,fontweight="bold")
ax.hlines(y=acc_bottom["pathwayid"], xmin=0, xmax=16, color='gray', alpha=0.7, linewidth=2, linestyles='dashdot')
ax.scatter(y=acc_bottom["pathwayid"], x=-(np.log10(acc_bottom["Pvalue"])), s=300, color='firebrick', alpha=0.8, marker="s",label='Late')
ax.scatter(y=acc_bottom["pathwayid"], x=-(np.log10(acc_top["Pvalue"])), s=300, color='grey', alpha=0.8, marker="o",label='Early')
ax.yaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
l=ax.legend(loc="upper left",bbox_to_anchor=(1,1),frameon=True,title="Stage",title_fontsize=40,fontsize=40)
l.get_title().set_weight("bold")
plt.tight_layout()
plt.savefig('replication_acc_pval.svg')
#,prop={'size':22}
```
